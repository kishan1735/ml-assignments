```python
def visualize_column_distribution(column_name):  
    fig,axes = plt.subplots(1, 3, figsize=(18, 6))
    
    sns.kdeplot(housing_df[column_name], ax=axes[0], fill=False)
    axes[0].set_title(f'KDE of {column_name}')
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Density')
    
    sns.boxplot(data= housing_df, x=column_name, ax=axes[1])
    axes[1].set_title(f'Boxplot of {column_name}')
    axes[1].set_xlabel(column_name)
    
    sns.scatterplot(data=housing_df,x=column_name, y='median_house_value', ax=axes[2])
    axes[2].set_title(f'{column_name} vs median house value')
    axes[2].set_xlabel(column_name)
    axes[2].set_ylabel('median house value')

    plt.tight_layout()
    plt.show()
```

```python
visualize_column_distribution('total_bedrooms')
```

```python
def visualize_outliers_based_on_std(housing_df, column_name):    
    mean = housing_df[column_name].mean()
    std_dev = housing_df[column_name].std()
    
    lower_bound = mean - 2 * std_dev
    upper_bound = mean + 2 * std_dev
    
    outliers = housing_df[(housing_df[column_name] < lower_bound) | (housing_df[column_name] > upper_bound)]
    num_outliers = outliers.shape[0]
    
    print(f"Total number of outliers in '{column_name}': {num_outliers}")
    
    plt.figure(figsize=(12, 6))
    plt.hist(housing_df[column_name], bins=50, color='blue', alpha=0.5, edgecolor='black')
    plt.axvspan(housing_df[column_name].min(), lower_bound, color='red', alpha=0.3)
    plt.axvspan(upper_bound, housing_df[column_name].max(), color='red', alpha=0.3)
    plt.title(f'Outlier Detection in {column_name} (95% Interval)')
    plt.xlabel(column_name)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

```python
visualize_outliers_based_on_std(housing_df, 'median_house_value')
```

### Resulting ranges

- **Min Max Scaling** - Transforms features to range [0, 1].

- **Standard Scaling** - Transforms features to have a mean of 0 and standard deviation of 1

### Min Max Scaling

**Benefits:**
- All features are on the same scale
- Simple to implement and understand

**Drawbacks:**
-  Sensitive to outliers

### Standard Scaling

**Benefits:**
- Not affected by outliers as much as Min-Max scaling, since it focuses on the mean and standard deviation
- Suitable for algorithms that assume normal distribution

**Drawbacks:**
- Extreme outliers can affect

### More Methods

- **RobustScaler** - ideal for data with many outliers.

- **QuantileTransformer** - to transform the data to a uniform or normal distribution.

- **PowerTransformer** - make data more Gaussian-like, mainly for models assuming normality.
